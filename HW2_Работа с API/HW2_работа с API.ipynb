**Уточнение.** Вы можете решать задачу любым способом, главное, получить список нужного вида. Можно обойтись одним циклом, а можно написать функцию для получения списка с информацией по одному актёру, а дальше применить её ко всем элементам списка `act_info`.

````python
### YOUR CODE HERE ###
def parse_actor(card):
    # id роли
    try:
        role_id = card.find("div", class_="actor_film_descript")["id"]
    except Exception:
        role_id = ""

    # ФИО
    try:
        name = card.find("strong", attrs={"itemprop": "name"}).get_text(strip=True)
    except Exception:
        name = ""

    # ссылка
    try:
        link = card.find("div", class_="film_name").find("a")["href"]
    except Exception:
        link = ""

    # роль
    try:
        role = card.find("div", class_="film_role").get_text(" ", strip=True)
    except Exception:
        role = ""

    return [role_id, name, link, role]

actors_data = [parse_actor(card) for card in act_info]

print(f"Статус ответаe: {page.status_code}")
print(f"Количество актеров в списке: {len(actors_data)}\n")

for item in actors_data:
    print(item)
````
